5. Write a function random_array(size, minVal=0, maxVal=1) which returns a NumPy
array of the given size containing random numbers in the interval given by the default
arguments minVal and maxVal .

```python
def random_array(size, minVal=0, maxVal=1):
    return np.random.uniform(minVal, maxVal, size)
my_array = random_array((3, 4), 0, 10)
print(my_array)
```

6. Write a function analyze(array) which prints out the minimum , maximum , mean , and 
median values of the provided array and returns a dictionary containing these values under
appropriate keys.

```python
def analyze(array):
    min_val = np.min(array)
    max_val = np.max(array)
    mean_val = np.mean(array)
    median_val = np.median(array)
    print(f"Minimum: {min_val}")
    print(f"Maximum: {max_val}")
    print(f"Mean: {mean_val}")
    print(f"Median: {median_val}")
    return {"minimum": min_val, "maximum": max_val, "mean": mean_val, "median": median_val}
result = analyze(np.array([1, 5, 2, 4, 3]))
print(result)
```

7. Write a function trigonometric_plot(x, plot_type) , which takes in a NumPy array x
and gives the plot of either sin(x) , cos(x) or tan(x) depending on the value assigned to
the plot_type argument. Your function should use the Matplotlib library.

```python
import matplotlib.pyplot as plt
def trigonometric_plot(x, plot_type):
    if plot_type == 'sin':
        y = np.sin(x)
        plt.plot(x, y)
        plt.title("Sin(x) Plot")
    elif plot_type == 'cos':
        y = np.cos(x)
        plt.plot(x, y)
        plt.title("Cos(x) Plot")
    elif plot_type == 'tan':
        y = np.tan(x)
        plt.plot(x, y)
        plt.title("Tan(x) Plot")
    else:
        print("Invalid plot type. Choose 'sin', 'cos', or 'tan'.")
        return
    plt.xlabel("x")
    plt.ylabel(plot_type)
    plt.show()
my_array = np.linspace(-5, 5, 75)
trigonometric_plot(my_array, 'sin')
trigonometric_plot(my_array, 'cos')
```

9. Compute the BMI for each pair of data-points from Q.8 above. Add the BMI values as a new
column in the existing Pandas DataFrame. BMI = (weight in kgs)/(height in
meters)^2 (Hint: Be careful with the units!)

```python
data = pd.read_csv('weight_height_data.csv')

data['Height'] = data['Height'] / 39.37
data['Weight'] = data['Weight'] * 0.453592

data['BMI'] = data['Weight'] / (data['Height'] ** 2)

plt.scatter(data['Height'], data['Weight'])
plt.xlabel('Height (meters)')
plt.ylabel('Weight (kilograms)')
plt.title('Height vs Weight (BMI included)')
plt.show()
```

10. Finally, compute the average BMI using the data from Q.9 above

```python
data = pd.read_csv('weight_height_data.csv')

data['Height'] = data['Height'] / 39.37
data['Weight'] = data['Weight'] * 0.453592

data['BMI'] = data['Weight'] / (data['Height'] ** 2)

avg_bmi = data['BMI'].mean()
print('Average BMI: ', avg_bmi)
```
